## Objective

Identify the different experiment types available in the NASA Battery Dataset.

The dataset contains three experiment categories:

- Charge
- Discharge
- Impedance

## Dataset Size Analysis

```python
rows = count_rows(files)

print(f"Minimum rows : {min(rows)}")
print(f"Maximum rows : {max(rows)}")
print(f"Average rows : {sum(rows)/len(rows):.2f}")
```

## Column Format Identification

```python
column_formats = {}

for file in files:

    df = pd.read_csv(file)

    columns = tuple(df.columns)

    if columns not in column_formats:
        column_formats[columns] = []

    column_formats[columns].append(file.name)
```

```python
summary = []

for i, (columns, files) in enumerate(column_formats.items(), start=1):

    summary.append({
        "Format": i,
        "Number_of_Files": len(files),
        "Example_File": files[0],
        "Columns": ", ".join(columns)
    })

summary_df = pd.DataFrame(summary)

summary_df
```

## Experiment Type Distribution

```python
for i, (columns, files) in enumerate(column_formats.items(), start=1):

    print("="*70)

    print(f"FORMAT {i}")

    print("="*70)

    print()

    print("Columns:")

    for c in columns:
        print("-", c)

    print()

    print("Number of files :", len(files))

    print("Example file :", files[0])

    print()
```
